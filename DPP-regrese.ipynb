```python
import pandas as pd
from pandas.stats.api import ols
```

```python
data = pd.read_csv('data/processed/delayed.csv')
```

```python
ols(y=data['delay'], x=data[['a','b','c','d','e','f','g']])
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula="delay ~ a + b + c + d + e + f + g", data=data).fit()
result.summary()
```
